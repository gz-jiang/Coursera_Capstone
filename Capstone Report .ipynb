### RESULTS

During the analysing of light conditions, it is a little bit suprised that the major part of accidents happened in day time, which could provise a better and clearer view. However, in the night time, it is very obvious that a good light condition could avoid a car accident in most cases. When drivers dirved their cars in the lit light or even no light, the probability of accidents' happening is vastly greater than the situation with a good light conditions. 

Secondly, we could notice that on single carriageway, the amount of car accidents is bigger than the situation on any other types or road. In fact, the single carriageway's accidents took almost three fourth of all the accidents. Of course, a large amount of single carriageway could also become the reason of this phenomenon. 

Another interesting finding is that the accidents happened on Saturday is obvious more than the weekday-accidents while the accidents happened on Sunday is obvious less than the situation in Saturday. In my opinion, as the first day in the weekend, maybe many people would choose to go out on Saturday, such as do some shopping or have a short trip with their families and friends; but on Sunday, people may prefer to stay at home for resting because tomorrow they need to go to work again. On the other hand, the accident happened on Monday is the least. I think this is because on that day people have already rested well during the weekend, so they would not feel too tired and they could pay more attention on driving. This might also be another reason of why Saturday's accidents is most, after working for a whole week, it is reasonable that people are too tired to drive safety.
